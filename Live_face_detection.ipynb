```python
import cv2
import numpy as np
import tensorflow as tf
from keras import models

face_cascade = cv2.CascadeClassifier('C:\\Users\\glads\\PycharmProjects\\pythonProject\\haarcascade_frontalface_default.xml')
yawn_detect = models.load_model('C:\\Users\\glads\\PycharmProjects\\pythonProject\\model.keras')
```

```python
feed = cv2.VideoCapture(0)
while True:
    ret, frame = feed.read() # frame RGB video feed
    
    dim = (320,240)
    resized = cv2.resize(frame, dim, interpolation = cv2.INTER_AREA)
    gray_resized = cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY)
    gray = cv2.equalizeHist(gray_resized)

    X = np.array(gray)
    X = X.reshape((1,240,320,1))
    yawn = np.argmax(yawn_detect.predict(X),axis=1)

    faces = face_cascade.detectMultiScale(gray)   
    if len(faces) != 0:
        for (x,y,w,h) in faces:
            if yawn == 0:
                print("Tired")
                cv2.rectangle(resized,(x,y),(x+w,y+h),(255,255,255),2)
            else:
                print("Not Tired")
                cv2.rectangle(resized,(x,y),(x+w,y+h),(0,0,0),2)
    
    cv2.imshow('video',resized)    
    if cv2.waitKey(1) == 27: # 27 is esc button
        break

feed.release()
cv2.destroyAllWindows()
```
